```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv("Resources/athletic_sales_2020.csv")
athletic_sales_2021_df = pd.read_csv("Resources/athletic_sales_2021.csv")
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.

athletic_sales_2020_21_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df]).reset_index()
athletic_sales_2020_21_df
```

```python
# Check if any values are null.
athletic_sales_2020_21_df.isnull().sum()
```

```python
# Check the data type of each column
athletic_sales_2020_21_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic_sales_datetime_df = athletic_sales_2020_21_df.copy()
athletic_sales_datetime_df["invoice_date"] = pd.to_datetime(athletic_sales_datetime_df["invoice_date"], format="mixed")
athletic_sales_datetime_df
```

```python
# Confirm that the "invoice_date" data type has been changed.

athletic_sales_datetime_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
athletic_sales_total_products_df = athletic_sales_2020_21_df.groupby(by=['region','state','city']).agg('sum').reset_index()
athletic_sales_total_products_df = athletic_sales_total_products_df.sort_values(['units_sold'], ascending=False)
athletic_sales_total_products_df = athletic_sales_total_products_df.rename(columns={'units_sold':"Total_Products_Sold"})
athletic_sales_total_products_df = athletic_sales_total_products_df[['region','state','city','Total_Products_Sold']]

# Show the top 5 results.
athletic_sales_total_products_df.head(5)

```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_athletic_sales_total_products_df = pd.pivot_table(athletic_sales_2020_21_df,
                                                  index=['region','state','city'], 
                                                  values='units_sold', 
                                                  aggfunc='sum').reset_index()                                           
                                          

pivot_athletic_sales_total_products_df = pivot_athletic_sales_total_products_df.sort_values(['units_sold'], ascending=False)
pivot_athletic_sales_total_products_df = pivot_athletic_sales_total_products_df.rename(columns={'units_sold':"Total_Products_Sold"})
pivot_athletic_sales_total_products_df = pivot_athletic_sales_total_products_df[['region','state','city','Total_Products_Sold']]
                                                                                                  
# Rename the "units_sold" column to "Total_Products_Sold"
# Show the top 5 results.
pivot_athletic_sales_total_products_df.head(5)

```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
woman_atletic_footware = athletic_sales_2020_21_df[athletic_sales_2020_21_df["product"] == "Women's Athletic Footwear"].copy()
woman_atletic_footware
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
woman_atletic_footware_total_sales_df = woman_atletic_footware.groupby(by=['retailer', 'region','state','city']).agg('sum').reset_index()
woman_atletic_footware_total_sales_df = woman_atletic_footware_total_sales_df.sort_values(['units_sold'], ascending=False)
woman_atletic_footware_total_sales_df = woman_atletic_footware_total_sales_df.rename(columns={'units_sold':"Womens_Footwear_Units_Sold"})
woman_atletic_footware_total_sales_df = woman_atletic_footware_total_sales_df[['retailer','region','state','city','Womens_Footwear_Units_Sold']]

# Show the top 5 results.
woman_atletic_footware_total_sales_df.head(5)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_woman_athletic_footware_total_sales_df = pd.pivot_table(woman_atletic_footware,
                                                  index=['retailer','region','state','city'], 
                                                  values='units_sold', 
                                                  aggfunc='sum').reset_index()                                           
                                          

pivot_woman_athletic_footware_total_sales_df = pivot_woman_athletic_footware_total_sales_df.sort_values(['units_sold'], ascending=False)
pivot_woman_athletic_footware_total_sales_df = pivot_woman_athletic_footware_total_sales_df.rename(columns={'units_sold':"Womens_Footwear_Units_Sold"})
pivot_woman_athletic_footware_total_sales_df = pivot_woman_athletic_footware_total_sales_df[['retailer','region','state','city','Womens_Footwear_Units_Sold']]

# Show the top 5 results.
pivot_woman_athletic_footware_total_sales_df.head(5)
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
# Optional: Rename the "total_sales" column to "Total Sales"
woman_athletic_footware_sales_bydate_df = pd.pivot_table(woman_atletic_footware,
                                                  index='invoice_date', 
                                                  values='total_sales', 
                                                  aggfunc='sum').reset_index()
woman_athletic_footware_sales_bydate_df["invoice_date"] = pd.to_datetime(woman_athletic_footware_sales_bydate_df["invoice_date"], format="mixed")                                         
woman_athletic_footware_sales_bydate_df = woman_athletic_footware_sales_bydate_df.sort_values(['invoice_date'], ascending=True)                                        
woman_athletic_footware_sales_bydate_df = woman_athletic_footware_sales_bydate_df.rename(columns={'total_sales':"Total Sales"})
woman_athletic_footware_sales_bydate_df = woman_athletic_footware_sales_bydate_df[['invoice_date','Total Sales']]

# Show the top 5 results.
print(woman_athletic_footware_sales_bydate_df[['invoice_date','Total Sales']].to_string(index=False))
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
# Sort the resampled pivot table in descending order on "Total Sales".
woman_athletic_footware_daily_sales_df = woman_athletic_footware_sales_bydate_df.resample('D',on='invoice_date').sum().reset_index()                                                                                                                       
woman_athletic_footware_daily_sales_df = woman_athletic_footware_daily_sales_df.sort_values(['Total Sales'], ascending=False)


# Show the top 5 results.
print(woman_athletic_footware_daily_sales_df[['invoice_date','Total Sales']].to_string(index=False))

```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
# Sort the resampled pivot table in descending order on "Total Sales".
woman_athletic_footware_weekly_sales_df = woman_athletic_footware_sales_bydate_df.resample('W',on='invoice_date').sum().reset_index()                                                                                                                       
woman_athletic_footware_weekly_sales_df = woman_athletic_footware_weekly_sales_df.sort_values(['Total Sales'], ascending=False)

print(woman_athletic_footware_weekly_sales_df[['invoice_date','Total Sales']].to_string(index=False))
```
